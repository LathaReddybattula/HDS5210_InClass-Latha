```python
import pandas as pd
import datetime
```

```python
ndc = pd.read_csv('//content/product (1).txt',delimiter='\t',dtype=str)
```

```python
ndc.head()
```

```python
cert_through = ndc['LISTING_RECORD_CERTIFIED_THROUGH'].astype(str)
```

```python
end_date = pd.to_datetime(cert_through.str[0:8], format='%Y%m%d')
```

```python
end_date.head()
```

```python
ndc['end'] = end_date
```

```python
market_start = ndc['STARTMARKETINGDATE'].astype(str)
```

```python
ndc['start'] = pd.to_datetime(market_start.str[0:8], format='%Y%m%d')
```

```python
ndc.head(100)
```

# Date Differences


```python
# Time on market
# Calculate the difference in days
time_on_market = (ndc['end'] - ndc['start']).dt.days

# Divide by 365.25 to get an approximate number of years
time_on_market_years = time_on_market / 365.25
```

```python
ndc.head()
```

```python
import numpy as np

ndc['duration'] = ndc['end'] - ndc['start']
# Calculate the difference in days
ndc['days'] = ndc['duration'].dt.days
# Divide by 365.25 to get an approximate number of years
ndc['years'] = ndc['days'] // 365.25
```

```python
ndc.head()
```

```python
ndc.shape
```

```python
ndc['age rank'] = ndc.years.rank()
```

```python
ndc.head()
```
